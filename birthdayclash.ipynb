```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
## load the birth data
bdata = pd.read_csv("http://chmullig.com/wp-content/uploads/2012/06/births.csv")
bdata.to_csv('births.csv')
```

```python
## days with fewer than 500 are bad data
bdata = bdata[bdata.births>500]
```

```python
## build up a series with all the day numbers in the right ratio

daynum = 1
birthdays = []

for day in bdata.values:
    for i in range( 0, day[2]):
        birthdays.append(daynum)
        
    daynum = daynum + 1
    
bday_series = pd.Series(birthdays)
```

```python
sims = 10000000      # pick a number of simulations
np.random.seed = 686 # make the outcome repeatable
simresult = pd.Series(np.zeros([sims], dtype=int)) # initialize a series to put results in
```

```python
%%time 
## on my MBP it takes about 1 second per 1000 sims
loopstart = dt.datetime.now()                       # initial start time

for i in range( 0, sims ):                          # do this for each sim
    sample = bday_series.sample( 7, replace=True )  # grab 7 dates from the emperical series
    simresult[i] =  max(sample.value_counts())      # grab the max from value_count; the day with the most clashes
    if i%10000 == 0:                                # these lines just print out loop number and run time
        print str( i )  + ' ' + str(  dt.datetime.now() - loopstart )
        loopstart  = dt.datetime.now()
    
```

```python
pd.Series(simresult).value_counts() / sims
```
